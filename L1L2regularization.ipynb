```python
dataset = pd.read_csv('/content/drive/MyDrive/ML/Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
dataset.head()
```

```python
X = dataset.drop('Price', axis = 1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y)
```

```python
reg.score(train_X,train_y)
```

```python
reg.score(test_X, test_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter = 100, tol=0.1)
lasso_reg.fit(train_X,train_y)
```

```python
reg.score(test_X,test_y)
```

```python
reg.score(train_X,train_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = linear_model.Ridge(alpha=50,max_iter = 100, tol=0.1)
ridge_reg.fit(train_X,train_y)
```

```python
ridge_reg.score(test_X,test_y)
```

```python
ridge_reg.score(train_X,train_y)
```
